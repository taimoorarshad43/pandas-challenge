## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Remember, theres repeat data, so find a way to only find the unique users and their
# respective genders
# Apparently, dataframe.loc[] will give duplicates

# Strategy: use dictionaries (again)

# Need to generate the final dataframe

gen = {}

namegender = df.loc[:,['SN','Gender']]

namegender

for x in range(len(namegender)):
    snname = namegender.iloc[x,0]
    
    if snname not in gen:
        gen[snname] = namegender.iloc[x,1]

totalgenders = len(gen)

males = 0
females = 0
others = 0

for x in gen:
    if gen[x] == "Male":
        males += 1
    elif gen[x] == "Female":
        females += 1
    elif gen[x] == "Other / Non-Disclosed":
        others += 1
    else:
        continue

gen


# Answers: Count and Percents of Genders (unique)
percentmale = males/totalgenders * 100
percentfemale = females/totalgenders * 100
percentothers = others/totalgenders * 100

# print(males)
# print(females)
# print(others)

genderslist = [males, females, others]

genderspercents = [percentmale, percentfemale, percentothers]

overallgenders = ["Male", "Female", "Other / Non-Disclosed"]

df_finalgender = pd.DataFrame({"Genders" : overallgenders, "Total Count": genderslist
                              , "Percentage of Players": genderspercents})

df_finalgender = df_finalgender.set_index(['Genders'])

df_finalgender['Percentage of Players'] = df_finalgender['Percentage of Players'].map("{:.2f}%".format)


df_finalgender

# print(percentmale)
# print(percentfemale)
# print(percentothers)


# Debug Statements
# ~~~~~~~~~~~~~~~~~

# len(gen)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# This seems to be a groupby section

# Average total per person by gender seems to be an issue. Can't just be done by groupby.
# It seems to want unique person values and then their purchasing average grouped by gender.
# The unique part seems to be an issue but maybe there's a method that can help with that. But what?

# What about groupby but by Gender and SN?

# Need to Generate the final Dataframe




gender_groupby = df.groupby(['Gender'])

gender_df = pd.DataFrame(gender_groupby.sum())

# Answers: For Genders (unique) Average Purchase, Purchase Count, and Total Purchase
avgpurgen = gender_groupby['Price'].mean()
countpurgen = gender_groupby['SN'].count()
totpurgen = gender_groupby['Price'].sum()


# Answer: Average Purchases by Gender by Person
# And this works!

gendertotals = []

for x in gender_df['Price']:
    gendertotals.append(x)
    
gendertotals

gendertotals[0] = gendertotals[0]/females

gendertotals[1] = gendertotals[1]/males

gendertotals[2] = gendertotals[2]/others

gendertotals



gender_financial_final = pd.DataFrame(gender_groupby['Purchase ID'].count())

gender_financial_final['Average Purchase Count'] = avgpurgen

gender_financial_final['Total Purchase Value'] = totpurgen

gender_financial_final['Avg Total Purchase per Person'] = gendertotals

gender_financial_final = gender_financial_final.rename(columns = {"Purchase ID" : 'Purchase Count'})

gender_financial_final

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Again you need unique data

# So far a winning strategy involves using dictionaries and making a dataframe out of the 
# the data in the dictionary

# Need to drop the SN column

bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]
labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']


sn_df = df.loc[:,['SN', 'Age']]

sn_age = {}

for x in range(len(sn_df)):
    
    sn = sn_df.iloc[x,0]
    age = sn_df.iloc[x,1]
    
    if sn not in sn_age:
        sn_age[sn] = age
        
len(sn_age)


names = []

ages = []

for x in sn_age:
    
    names.append(x)
    ages.append(sn_age[x])
    
# Answer: Age groups (unique)    
names_age_df = pd.DataFrame({"SN": names, "Age" : ages})

names_age_df["Age Group"] = pd.cut(names_age_df['Age'], bins, labels = labels)

names_age_df.head(576)

agegroup = names_age_df.groupby(['Age Group'])

agegroup.count().head(10)

agegroup_df = pd.DataFrame(agegroup.count())


# Answer: Age Group (unique) Percentages
agegroup_df['Percents'] = agegroup_df['Age']/len(sn_age) * 100

del agegroup_df["SN"]

agegroup_df = agegroup_df.rename(columns = {'Age' : 'Total Count', 'Percents': 'Percentage of Players'})

agegroup_df["Percentage of Players"] = agegroup_df["Percentage of Players"].map("{:.2f}%".format)

agegroup_df
```

```python
# Seems like groupby but for SN
# You need to be able to sort through a column that accounts for all the values given not just count
# or average or total but all three at the same time

dfcopy = df.copy()

sngb = dfcopy.groupby(['SN'])

# This will be a column of the final dataframe
sngb[['Purchase ID']].count().head()

# This will be a column of the final dataframe
sngb[['Price']].mean().head()

# This will be a column of the final dataframe
sngb[['Price']].sum().head()

df_sorted = pd.DataFrame()

df_sorted['Purchase Count'] = sngb['Purchase ID'].count()

df_sorted['Average Purchase Price'] = sngb['Price'].mean()

df_sorted['Total Purchase Value'] = sngb['Price'].sum()

df_sorted.head()

# Answer: Sorted by Purchase Count, Average Purchase Price, and Total Purchase Value
df_finalsorted = df_sorted.sort_values(['Total Purchase Value']
                                      , ascending = False)

df_finalsorted.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Sort purchase count column in descending order

item_idnameprice = df.loc[:, ['Item ID','Item Name', 'Price']]

item_idnameprice

items_groupby = item_idnameprice.groupby(['Item ID', 'Item Name'])

# Answer: Item count, to be sorted
items_groupby.count().head(10)

# Answer: Item price, to be sorted
items_groupby.mean().head(10)

# Answer: Total Item Price, to be sorted
items_groupby.sum().head(10)

df_mostpop = pd.DataFrame(items_groupby.count())

df_mostpop = df_mostpop.rename(columns = {'Price': 'Purchase Count'})

df_mostpop['Item Price'] = items_groupby.mean()

df_mostpop['Total Purchase Value'] = items_groupby.sum()

df_mostpop = df_mostpop.sort_values('Purchase Count', ascending = False)

df_mostpop.head()
```
